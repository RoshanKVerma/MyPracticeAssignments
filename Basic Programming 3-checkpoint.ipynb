1.	Write a Python Program to Check if a Number is Positive, Negative or Zero ?

```python
t=float(input('Enter the Number to Check'))
if t==0:
    print('The Number is Zero, Entered Number=',t)
elif t > 0:
    print('The Number is Positive, Entered Number=',t)
elif t < 0:
    print('The Number is Negetive, Entered Number=',t)
else:
    print('Error in input Number')
```

2.	Write a Python Program to Check if a Number is Odd or Even ?

```python
n=float(input('Enter The Number'))
if n%2==0:
    print('The number is Even')
else:
    print('The number is odd')
```

3.	Write a Python Program to Check Leap Year ?

```python
y=int(input('Enter the Year'))
if y%4==0:
    print('This year is Leap Year')
else:
    print('This year is not a Leap Year')
```

4.	Write a Python Program to Check Prime Number ?

```python
p=int(input('Input the number to check'))
if p > 1:
    for i in range(2,p-1):
        if p%i==0:
            print('The number is not Prime Number')
            break
        else:
            print('The number Prime number')
            break
else:
    print('Number is less than 1')
```
